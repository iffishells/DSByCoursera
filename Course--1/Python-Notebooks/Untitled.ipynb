## Hyphotheses testing 

```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df  = pd.read_csv('datasets/grades.csv')
df.head()
```

```python
df.shape
```

```python
early_finishers = df[pd.to_datetime(df["assignment1_submission"]) < '2016']
```

```python
early_finishers
```

```python
late_finishers = df[~df.index.isin(early_finishers.index)]
```

```python
late_finishers
```

```python
print(early_finishers["assignment1_grade"].mean())
print(late_finishers["assignment1_grade"].mean())
```

```python
## seems to be simirlar but not actual in order to this similar or not we
## we used here hyphothesis testing
```

```python
from scipy.stats import ttest_ind

ttest_ind(early_finishers["assignment1_grade"] , late_finishers["assignment1_grade"])
```

```python
## for the shake of understands lets take two same list
```

```python
import numpy as np
list_1 = [1,23,4,4,5]
list_2 = [1,23,4,4,5]
print(np.mean(list_1))
print(np.mean(list_2))
ttest_ind(list_1 , list_2)
```
